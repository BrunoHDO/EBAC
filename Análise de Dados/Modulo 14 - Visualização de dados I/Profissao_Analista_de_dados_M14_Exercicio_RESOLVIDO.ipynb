## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```
# gráfico do exercício 1

grafico = sns.barplot(data=diamonds, x='cut', y='price', palette='bright',order=['Premium', 'Fair', 'Very Good', 'Good', 'Ideal' ])

grafico.set(title='Preço do Diamante por Tipo de Corte',
            ylabel='Preço',
            xlabel='Tipos de Cortes',
            ylim = [0,5000]);

grafico.set_title('Preço do Diamante por Tipo de Corte', fontsize=14, fontweight='bold', pad= 15)
grafico.set_xlabel('Tipos de Cortes', fontsize=14)
grafico.set_ylabel('Preço', fontsize=14)
```

**Insight do gráfico 1**: Os tipos de cortes parecem ter uma relação com o preço do diamante. Os diamantes com corte Premium e Fair apresentam o maior valor. Enquanto o Very Good e Good, visualmente, apresentam valores bem similares, com uma diferença mínima. Curiosamente, o tipo de corte Ideal atribui um valor menor aos demais. ## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

flights = sns.load_dataset("flights")
flights.head()
```

```
# gráfico do exercício 2
import numpy as np

flights = flights.query('month == "Dec"')

with sns.axes_style('whitegrid'):
  grafico = sns.lineplot (data=flights, x='year', y='passengers')
  grafico.set(title = 'Número de passageiros em dezembro por ano',
                      xlabel = 'Ano',
                      ylabel = 'Passageiros',
                      ylim = [100, 500])

grafico.set_title('Número de passageiros em dezembro por ano', fontsize=14, fontweight='bold', pad= 15)
grafico.set_xlabel('Ano', fontsize=14)
grafico.set_ylabel('Passageiros', fontsize=14)

grafico.grid(alpha=0.35)
```

**Insight do gráfico 2**: O número de passegeiros em dezembro, por ano, parece aumentar ao longo do tempo. Econtra-se apenas uma exceção entre os anos 1957-1958, onde o número de passageiros foi igual para o mês de dezembro de cada ano.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query('year >= 1949 and year <= 1959')

with sns.axes_style('whitegrid'):
  cor = sns.color_palette("viridis", as_cmap=True)
  grafico = sns.lineplot(data=flights, x='month', y='passengers', hue='year', palette= cor )

  grafico.set(title = 'Número de passaeiros por mês',
                      xlabel = 'Mês',
                      ylabel = 'Passageiros')

  grafico.set_title('Número de passaeiros por mês', fontsize=14, fontweight='bold', pad= 15)
  grafico.set_xlabel('Mês', fontsize=14)
  grafico.set_ylabel('Passageiros', fontsize=14)

  grafico.grid(alpha=0.35)
  grafico.legend(title='Ano',loc=[1.01,0])
```

**Insight do gráfico 3**:  Os anos apresentam uma mesma tendência quanto ao número de passageiros ao mês. Janeiro, fevereiro, maio e novembro são meses de baixa. Já os meses de março, agosto e dezembro são meses de alta.
